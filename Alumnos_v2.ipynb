```python
import pandas as pd
```

```python
url='https://gist.githubusercontent.com/ahcamachod/807a2c1cf6c19108b2b701ea1791ab45/raw/fb84f8b2d8917a89de26679eccdbc8f9c1d2e933/alumnos.csv'
datos=pd.read_csv(url)
```

```python
datos.info()
```

##**Los estudiantes participaron en una actividad extracurricular y ganaron puntos extras. Estos puntos extras corresponden al 40% de su nota actual. Por lo tanto, crea una columna llamada "Puntos_extras" que contenga los puntos extras de cada estudiante, es decir, el 40% de su nota actual.**

```python
datos['Puntos_extras']=datos['Nota']*.40
datos.head()
```

##**Crea otra columna llamada "Notas_finales" que contenga las notas de cada estudiante sumadas con los puntos extras.**

```python
datos['Notas_finales']=datos['Nota']+datos['Puntos_extras']
datos.head()
```

###**Dado que hubo una puntuacin extra, algunos estudiantes que no haban sido aprobados antes pueden haber sido aprobados ahora. En funcin de esto, crea una columna llamada "Aprobado_final" con los siguientes valores:**
####True: si el estudiante est aprobado (la nota final debe ser mayor o igual a 7.0).
####False: si el estudiante est reprobado (la nota final debe ser menor que 7.0).

```python
datos['Aprobado_final'] = datos['Notas_finales'] >= 7.0
datos.head()
```

###**Realiza una seleccin y verifica qu estudiantes no haban sido aprobados anteriormente, pero ahora fueron aprobados despus de sumar los puntos extras.**

```python
datos['Aprobado'] = datos['Aprobado'].astype(str).str.lower().map({'true': True, 'false': False, 'verdadero': True}).fillna(False).astype(bool)

nuevos_aprobados = datos[(datos['Aprobado'] == False) & (datos['Aprobado_final'] == True)]
nuevos_aprobados.head()
```

```python
datos=nuevos_aprobados
```
